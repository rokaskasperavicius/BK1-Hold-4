```python
import pandas as pd
from numpy.random import randn
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
u = 22.6
s = 2.4

n = 10

data = pd.Series(randn(n) * s + u)

print(data.sort_values())

print(data.mean())
print(data.std())
```

```python
data.plot.box()
```

```python
# Exercise 10.9.2

n_list = [10, 100, 1000]
u = 22.6
s = 2.4
experiment_count = 100

for n in n_list:
    mean_list = pd.DataFrame(s * randn(experiment_count, n) + u).mean(axis ='columns')
    
    # Plotting time
    plt.figure(figsize=[10, 5])
    plt.title(f'Mean spread with n = {n}')
    
    plt.hist(mean_list, bins = np.arange(20, 25, 0.2))
    
    plt.xlabel('Mean')
    plt.ylabel('Frequency')
    
    plt.show()
    
    
```
