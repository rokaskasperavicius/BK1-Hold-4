```python
from numpy.random import binomial
from scipy.special import binom

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Exercise 9.6

```python
n = 4
p = 1 / 6

number_of_threes = binomial(n, p, size=100)

df_dice = pd.DataFrame({
    'Threes': number_of_threes
})

df_frequency = pd.DataFrame({
    'Simulation': df_dice['Threes'].value_counts(sort=False, normalize=True)
})

# Theoretical probability

# Possible outcomes
k = df_frequency.index

# Binomial coefficient for each outcome
df_frequency['Theoretical'] = binom(n, k) * p**k * (1 - p)**(n - k)

# Plotting
df_frequency.plot.bar();

plt.title('Simulation vs Theoretical probability')
plt.ylabel('Relative frequency')
plt.xlabel('Number of threes when rolling a dice four times ')

plt.show()
```
