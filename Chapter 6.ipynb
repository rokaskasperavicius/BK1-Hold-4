```python
def find_a_b(x, y):
    if (len(x) != len(y)):
        raise Exception('Length mismatch!')
    
    a = x.cov(y) / x.var()
    
    b = y.mean() - a * x.mean()
    
    return [a, b]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
# Table 6.2

x = pd.Series([250, 300, 350, 470])
y = pd.Series([95, 120, 165, 170])

[a, b] = find_a_b(y, x)

print(f'Slope a = {a} and b = {b}')
```

```python
# Table 6.6

x = pd.Series([0.083, 0.115, 0.191, 0.299, 0.501, 0.696, 0.893, 1.084])
y = pd.Series([0, 1, 5, 10, 20, 30, 40, 50])

[a, b] = find_a_b(x, y)

print(f'Slope a = {a} and b = {b}')
```

### Exercise 6.3

```python
timber = pd.read_csv('timber.csv')
```

```python
x = timber['Diameter (in)']
y = timber['Volume (cubic feet)']

p = np.polyfit(x, y, 1)

a, b = p

print(f'The slope is a = {a}')
print(f'The intersection is b = {b}')

timber['fit'] = np.polyval(p, x)

timber['residuals'] = y - timber['fit']
```

```python
plt.figure(figsize = (10, 5))

plt.scatter(
    x = timber['Diameter (in)'],
    y = timber['Volume (cubic feet)'],
)

plt.plot(timber['Diameter (in)'], timber['fit'], '-')

plt.savefig('timber_best_fit.png')

plt.show()
```

```python
plt.figure(figsize = (10, 5))

plt.stem(timber['Diameter (in)'], timber['residuals'])

plt.savefig('timber_residuals.png')

plt.show()
```
