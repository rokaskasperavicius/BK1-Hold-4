### Exercise 10.7

```python
# Define custom normal distribution function to get values for the bell curve
def get_bell_curve_values(u, s):
    y = np.arange(0, 35, 0.1)
    N = np.sqrt(2 * np.pi) * s
    arg = -0.5 * (y - u)**2 / s**2
    f = np.exp(arg) / N
    
    return (y, f)
```

```python
N = 1000
u = 17 # Mean, mu

# Standart deviation list, sigma
s_list = [2, 4, 8]

# Loop through sigma list and show three different plots
for s in s_list:

    # Generating data with sigma value
    data = pd.Series(randn(N) * s + u)

    # Plotting
    plt.figure(figsize=[10, 5])
    plt.title(f'Normal Distribution with mean = {u} and sigma = {s}')

    plt.hist(data, density=True, bins=np.arange(0, 35, 1.0))

    (y, f) = get_bell_curve_values(u, s)

    plt.plot(y, f, 'r-')

    plt.xlabel('Meassured value')
    plt.ylabel('Distribution')
    
    plt.show()
```

From the graphs above you can notice that with a larger sigma value, the width of the normal distribution is also larger.

### Exercise 10.8

```python
popcorn_df = pd.read_csv('popcorn.csv')

longest_size = popcorn_df['Size (longest)']

u = longest_size.mean()
s = longest_size.std()

y = np.arange(0, 50, 0.1)
N = np.sqrt(2 * np.pi) * s
arg = -0.5 * (y - u)**2 / s**2
f = np.exp(arg) / N

# Plotting
plt.figure(figsize=[10, 5])
plt.title('Distribution of popcorn width')

plt.hist(longest_size, density=True)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.plot(y, f, 'r-')

plt.show()
```

### Exercise 10.10

```python
data = pd.Series([33, 26, 34, 31, 23, 25])

u = data.mean()
s = data.std()

# 10.10.1
print(f'Standart Error = {u / len(data)**0.5} \n')

# 10.10.2
t0975 = 2

SEy = s / len(data)**0.5

delta_mean = t0975 * SEy

print(f'The lower limit of the 95% confidence interval is {u - delta_mean}')
print(f'The upper limit of the 95% confidence interval is {u + delta_mean} \n')

# 10.10.3
t0835 = 1

SEy = s / len(data)**0.5

delta_mean = t0835 * SEy

print(f'The lower limit of the 67% confidence interval is {u - delta_mean}')
print(f'The upper limit of the 67% confidence interval is {u + delta_mean}')
```
